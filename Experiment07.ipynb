```python
# To print the factorial of a given number
n = int(input("Enter a number:"))
fact = 1
while n>0:
  fact *= n
  n -= 1
print("The Factorial =",fact)
```

```python
# To print the reverse of a number:
n = int(input("Enter a number:"))
rev = 0
while n>0:
   digit = n%10
   rev = rev*10 + digit
   n //= 10
print("The reversed number is:",rev)
```

```python
# To check if a given number is a plaindrome:
num = int(input("Enter a number:"))
temp = num
rev = 0
while num>0:
  rev = rev * 10 + num % 10
  num //= 10

if temp == rev:
  print("The number is a palindrome")
else:
  print("The number is not a palindrome")
```

```python
# To check if the word madam  is a palindrome or not:
s = "madam"
i,j =0,len(s) - 1
is_palindrome = True
while i<j:
  if s[i] != s[j]:
    is_palindrome = False
    break
  i += 1
  j -= 1
if is_palindrome:
  print("The word is a palindrome")
else:
  print("The word is not a palindrome")
```
